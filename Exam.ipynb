# 3 Iris 데이터 2차원으로 차원 축소 후 K = 2, 3, 4에 대해 군집화 해보고 성능 평가하기

```python
columns = ['sepal_length','sepal_width','petal_length','petal_width']
irisDF = pd.DataFrame(iris.data, columns=columns)
irisDF['target']=iris.target
irisDF.head(3)
```

```python
iris_scaled = StandardScaler().fit_transform(irisDF.iloc[:,:-1])
pca = PCA(n_components=2)

pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
pca_columns=['pca_component_1','pca_component_2']
irisDF_pca = pd.DataFrame(iris_pca, columns=pca_columns)
irisDF_pca['target']=iris.target
irisDF_pca
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300,
                random_state=2020).fit(irisDF_pca)

irisDF_pca['cluster'] = kmeans.labels_
```

```python
score_samples = silhouette_samples(iris_pca.data, irisDF_pca['cluster'])
print('silhouette_samples( ) return 값의 shape', score_samples.shape)

irisDF_pca['silhouette_coeff'] = score_samples

average_score = silhouette_score(iris_pca.data, irisDF_pca['cluster'])
print('붓꽃 데이터셋 Silhouette Analysis Score:{0:.3f}'.format(average_score))

irisDF_pca
```

```python
irisDF_pca.groupby('cluster')['silhouette_coeff'].mean()
```

### K = 2 일때 0.354445
### K = 3 일때 0.486002
### K = 4 일때 0.369907
